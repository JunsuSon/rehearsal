```python
# 리스트 선언하기
a = [] # 빈 리스트
b = [1,2,3] # 숫자 데이터를 담은 리스트
c = ['My','name','is','JY'] # 문자열 데이터를 담은 리스트
d = [1,2,'My','name'] # 숫자, 문자열 데이터를 담은 리스트
e = [1,2,['My','name']] # 이중 리스트 가능

print(a)
print(b)
print(c)
print(d)
print(type(e))
```

#### 리스트 데이터 접근 
 - 인덱스 번호 활용
 - 문자열 인덱싱, 슬라이싱 활용과 동일
 - 인덱싱: 무엇인가를 '가리킴'

```python
list1 = [2,5,7,9,10]
# 2 접근 
list1[0]
# 9 접근
list1[3] # list1[-2]
```

```python
# 슬라이싱 
# 리스트[시작:끝+1]
list3 = ['지수','문규','다인','수양','성문']
# 지수, 문규 접근 출력 하기 !
print(list3[0:2])
# 슬라이싱 결과는 여러개에 대해 리스트로 묶어서 반환
```

```python
# list.index(값) -> 인덱스 번호 반환하는 함수
# 입력
# 조건문
# 있다면 ~ (인덱스 번호 출력까지 index() 활용해보기 )
# 없다면 ~ 
lst = ['딸기','바나나','수박','체리','포도']
search = input('검색할 문자를 입력>> ')
if search in lst:
    # 있다면 ~ 
    print(f'{search}은 리스트에 {lst.index(search)+1}번째에 들어있습니다.')
else:
    print(f'{search}은 리스트에 들어있지 않습니다.')
```

```python
# 문자열대상.count(값)
# 문자열 변수 선언
s = 'Hi, My name is Mihee'
s = s.lower()
# 검색 문자 입력 받기
search = input('검색할 문자를 입력하세요: ')
cnt = 0
# 조건문
if search in s:
    cnt = s.count(search)
    print(f"'{search}'는 {cnt}번 들어 있어요.")
else:
    print(f"'{search}'는 들어있지 않습니다.")
    
```

### 문자열 접근하는 방법
 - 문자열 -> 데이터
 - 인덱싱, 슬라이싱

```python
# 인덱싱 -> 한자리만 가리킴
s = 'My name is JY'
print(s[0])
print(s[8])
print(s[-2])
print(s[-1])
```

```python
# 슬라이싱 -> 특정 문자열 접근
# [시작인덱스:끝인덱스+1]
# My 슬라이싱
s[0:2]

# name
s[3:7]
# is 
s[8:10:1]
```

```python
# 생략할 수 있는 인덱스 번호
# s 변수에서 맨 첫번째 인덱스부터 name e까지 포함해서 슬라이싱
print(s[0:7])
print(s[:7]) # 시작값의 기본값:0 (생략가능)
print(s[8:13])
print(s[8:]) # 끝값의 기본값: 문장의 끝 (생략가능)
print(s[:]) # 모든 문자열에 접근 # 증감값(step) :1
print(s[::2]) # 모든 문자열에 접근 2칸씩 뛰면서 접근
```

```python
# 길이 확인하는 함수
len(now)
now[10:len(now)]
```

```python
# 문자열 대소문자 변환 관련 함수
# 소문자를 대문자로 바꾸기
eng_str = 'hi, My name is JY'
eng_str2 = eng_str.upper() # 해당 변수에 초기화 x 
eng_str2
# 대문자를 소문자로 바꾸기 
eng_str3 = eng_str.lower()
eng_str3
```

```python
# '찾을 문자열'이 몇개 있는지 확인하는 함수 
# 'm' 카운트
# 대소 문자 구별 
eng_str.lower().count('m')
# .count('M')
```

```python
# 특정 문자열 찾아서 인덱스 번호 확인(반환)하는 함수
# find()
# index()
eng_str.find('i') # 제일 먼저 찾아지는 'i' 인덱스 값 출력 
eng_str.index('i')# 제일 먼저 찾아지는 'i' 인덱스 값 출력 

# 문자열에 없는 단어 찾아달라고 하기 
eng_str.find('z') # 제일 먼저 찾아지는 'i' 인덱스 값 출력 
# find 는 해당 단어가 없으면 -1 값 반환, 에러 x 
eng_str.index('z')# 제일 먼저 찾아지는 'i' 인덱스 값 출력 
# index는 해당 단어가 없으면 에러 발생 
```

```python
# strip() 문자열 양끝의 공백을 제거하는 함수 
'    배고프다!    '.strip()
```

```python
# 문자열1을 문자열2로 변경하는 함수
# replace(문자열1, 문자열2)
eng_str.replace('JY','JH')
```

```python
article = "[요기요, 화이트데이 맞이 '선물하기' 이벤트 진행] 기사입력 2022.03.10. 작성자 이나영기자 요기요는 화이트데이를 앞두고 요기요 선물하기 서비스를 이용하는 고객들을 위한‘너와 나의 선물고리’이벤트를 진행한다고 10일 밝혔다." 
# start 인덱싱 시작 정보
start = article.find('2022') 
# end 인덱싱 번호 끝 정보 
end = article.find(' 작성자')
# day 위의 변수 값을 활용해서 슬라이싱
day = article[start:end-1]
print(day)
print(f'기사가 업로드된 일자는 {day}입니다.')
```

```python
# dic2 -> 우리 반 친구들의 정보를 담은 딕셔너리 
class_info = {'name':['현석','수양','나행'],
             'age':[28,20,25],
             'blood':['B','A','O']}
class_info
```

```python
# 딕셔너리 for 활용
for key in class_info.keys():
    print(key)

# 현석이의 정보만 출력해보기( 이름, 나이, 혈액형)
for key in class_info.values():
    print(key[0])
```

```python
score_dict = {'이름':['손지영','김은영','김운비','채수민','최성우'],
        'Python':[90,100,85,90,80],
        'Java':[85,80,100,95,85],
        'html/css':[75,70,90,80,90]}
score_dict
```

```python
# 문제 1번. 파이썬 과목 평균 구하기
# 1. score_dict 파이썬 접근 후 
# 2. for 문 돌리면서 파이썬 성적 누적
# 3. 누적된 값 나눠서 평균 구하기 
py_total = 0
for score in score_dict['Python']:
    py_total += score
print(py_total/5)

# 추가) 리스트 안의 모든 값 더해주는 함수(총합 구하는 함수)
sum(score_dict['Python'])/5
```

```python
# 문제 2번. 각 과목의 평균 구하기 
# 1. 이름 빼고 나머지 과목 키값들만 접근하게 만들기 
score_dic_key = list(score_dict.keys())
score_dic_key[1:]
# 2. ['Python', 'Java', 'html/css'] 대해서 for문 돌리며
# 과목점수들에 접근하고 sum통해 총합, 평균 해결
avg = 0
for sub in score_dic_key[1:]:
    # score_dict 딕셔너리 , sub 키(과목명)
    avg = sum(score_dict[sub])/5
    print(f'{sub} 평균: {avg}')
```

```python
# '손지영' 사람의 데이터가 있어 
# for 문을 사람 이름 정보만큼 돌리면서 
# 손지영이라는 데이터는 몇번째 있었는지 인덱스 번호 찾는다!
score_dict['Python'][0] 
score_dict['Java'][0]
score_dict['html/css'][0]
```

```python
# 3번 round 함수 사용 법 round(실수값, 소수점자리)
# 소수점 둘째자리 까지 --> round(실수값, 2)

name = input('이름을 정확하게 기입하세요 >> ')
s_sum = 0


if (score_dict['이름'][0] == name):
    s_sum += score_dict['Python'][0]
    s_sum += score_dict['Java'][0]
    s_sum += score_dict['html/css'][0]
    print(f'{name}님의 전 과목 성적 평균: {round(s_sum/3,2)}')
    
elif (score_dict['이름'][1] == name):
    s_sum += score_dict['Python'][1]
    s_sum += score_dict['Java'][1]
    s_sum += score_dict['html/css'][1]
    print(f'{name}님의 전 과목 성적 평균: {round(s_sum/3,2)}')

elif (score_dict['이름'][2] == name):
    s_sum += score_dict['Python'][2]
    s_sum += score_dict['Java'][2]
    s_sum += score_dict['html/css'][2]
    print(f'{name}님의 전 과목 성적 평균: {round(s_sum/3,2)}')
    
elif (score_dict['이름'][3] == name):
    s_sum += score_dict['Python'][3]
    s_sum += score_dict['Java'][3]
    s_sum += score_dict['html/css'][3]
    print(f'{name}님의 전 과목 성적 평균: {round(s_sum/3,2)}')
    
elif (score_dict['이름'][4] == name):
    s_sum += score_dict['Python'][4]
    s_sum += score_dict['Java'][4]
    s_sum += score_dict['html/css'][4]
    print(f'{name}님의 전 과목 성적 평균: {round(s_sum/3,2)}')
    
else: 
    print('검색 결과가 없습니다.')
```

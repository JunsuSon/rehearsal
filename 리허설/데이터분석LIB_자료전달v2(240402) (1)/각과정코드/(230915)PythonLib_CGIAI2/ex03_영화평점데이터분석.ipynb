```python
import numpy as np
```

```python
# np.genfromtxt : txt 파일 불러오는 함수 
# dtype: 타입 설정
data = np.genfromtxt('./data/ratings.dat', delimiter="::", dtype = 'int32')
data
# 실수 -> 과학적 표기법 
# user_id, movie_id, rating, 시간(timestamp)
```

```python
user_ids = data[:,0]
user_ids

# 중복없이 확인해보자!
user_ids_unq = np.unique(user_ids)
user_ids_unq.size
# 사용자 아이디는 1~ 6040까지 있구나 ! 
```

```python
# 사용자 아이디가 1인 사람 모든 정보 접근하기 
# user_ids == 1
data[data[:,0] == 1][:,2].mean()
# data[data[:,0] == 1,2].mean()
```

```python
# 5000 user 평점 평균 구하기 
data[data[:,0] == 5000,2].mean()
```

```python
# 4800 user
data[data[:,0] == 4800,2].mean()
```

```python
# user_id_mean 빈 리스트 
# user_id : 중복없는 user 아이디 값들 
user_id = np.unique(data[:,0])
user_id_mean = []
for i in user_id:
    # np.mean(data[data[:,0] == i,2])
    user_mean = data[data[:,0] == i,2].mean()
    user_id_mean.append([i, user_mean])
user_id_mean
```

```python
# 평점 평균이 4점 이상인 id 
# boolean 인덱싱 
# user_id_mean arr로 변환
user_id_mean_arr = np.array(user_id_mean, dtype = 'int32' )
user_id_mean_arr
user_id_unq[user_id_mean_arr[:,1]>=4] 
#  4 이상인 True 에 해당하는 위치 id값만 출력 됨
```

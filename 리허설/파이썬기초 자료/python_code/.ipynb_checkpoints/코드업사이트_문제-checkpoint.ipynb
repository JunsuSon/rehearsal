#### 조건문 문제
- 2313 : 한라봉 포장0 

알바왕 비버가 한라봉 상자 포장 아르바이트를 한다.
상자의 크기는 용량에 따라 4종류(1kg, 3kg, 5kg, 10kg)로 나뉘는데, 반드시 상자의 용량만큼 한라봉을 담아야 한다.
즉, 상자의 용량보다 부족하거나 초과하여 한라봉을 담을 수 없다.
또한, 상자를 최소한으로 사용해서 한라봉을 포장해야 한다.
예를 들어, 3kg의 한라봉을 포장하기 위해서는 1kg 상자 3개가 아닌, 3kg 상자 1개를 사용해야 한다.
 




비버가 21kg의 한라봉을 모두 포장하려면, 최소 몇 개의 상자가 필요할까?
*본 문제는 주어진 문제에 대한 답을 출력하는 문제로서 출력 예시가 없다. 문제에 대한 답을 출력하면 된다.

```python
input_kg = 21
b_ten = b_five = b_thr = b_one = extra = b_total = 0
b_ten = input_kg//10
b_five = input_kg%10//5
b_thr = input_kg%10%5//3
extra = input_kg%10%5%3
b_one = extra if extra > 0 else b_one

print('[결과화면]')
print(f'10kg:{b_ten}개  5kg:{b_five}개  3kg:{b_thr}개  1kg:{b_one}개')
```

#### 조건 판별 문제
- 2319 : 감귤 나무 관리0

감귤왕 비버는 원형 감귤밭의 감귤 나무를 3종류로 구분하여 관리한다.
비버는 밭의 중심 좌표, 밭의 반지름, 각 감귤 나무의 좌표 데이터를 활용해서
밭 내부에 위치한 나무(in), 밭 경계에 위치한 나무(on), 밭 외부에 위치한 나무(out)로 구분한다.
 


감귤밭의 중심 좌표가 (1, 1)이고 감귤밭의 반지름이 3일 때,
좌표 (3, 4)에 위치한 감귤 나무는 어떻게 구분될까?
*본 문제는 주어진 문제에 대한 답을 출력하는 문제로서 출력 예시가 없다. 문제에 대한 답을 출력하면 된다.

```python
# 피타고라스의 정의
# 좌표 사이 거리 계산 
# a제곱 = b제곱+c제곱
import numpy as np
# q1. 밭의 반지름이 2, 중심(0,0)이고 감귤나무좌표(1,0)이다. 내부(in)/경계(on)/외부(out) 구분해보아라.
center_x,center_y = (0,0)
tree_x,tree_y = (3,0)
distance = np.sqrt((center_x-tree_x)**2+(center_y-tree_y)**2)
if distance == 2:
    print(distance,'경계(on)입니다.')
elif distance < 2:
    print(distance,'내부(in)입니다.')
else:
    print(distance,'외부(out)입니다.')
```

```python
# q2. 밭의 반지름이 3, 중심(1,1)이고 감귤나무좌표(3,4)이다. 내부(in)/경계(on)/외부(out) 구분해보아라.
center_x,center_y = (1,1)
tree_x,tree_y = (3,4)
distance = np.sqrt((center_x-tree_x)**2+(center_y-tree_y)**2)
if distance == 3:
    print('on')
elif distance < 3:
    print('in')
else:
    print('out')
```

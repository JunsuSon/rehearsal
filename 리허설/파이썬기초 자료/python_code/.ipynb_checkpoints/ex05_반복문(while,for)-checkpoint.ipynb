####  반복문이란?
- 프로그램 내에서 똑같은 명령을 일정 횟수만큼 반복하여 수행하도록 제어하는 명령문
- 종류: while, for문

- while: 반복 횟수가 명확하지 않을때/ 주로 조건식과 같이 사용
- for: 반복횟수가 명확할때 / 주로 리스트,튜플,문자열에 대한 반복적인 실행이 필요할때 사용

#### 반복문이 필요한 이유
1~1000까지 출력

```python
print(1)
print(2)
print(3)
print(4)
print(5)
print(6)
# ...
```

```python
for i in range(1,1001):
    print(i)
```

#### while문
조건식이 True 일 경우 실행문장 반복


```python
number = 1
while number <= 3:
    print(number)
    number +=1
```

- while문 예제
while 문을 사용해서 '파이썬 최고 !!'13번 출력해보자.

```python
# while 조건식:
#      실행문장 
# 조건식이 true 값이면 while 안에 있는 실행문장 읽고 실행

num = 0 
while num <13:
    num +=1
    print(f'{num} 파이썬 최고 !! ')
```

- while문 break 키워드 !
반복을 멈춰라 !! 

```python
# while True:
#     print('어떻게 될까요 ?')

#-> 무한루프를 돔 ,,,,, 
```

```python
# break를 한번 넣어보자 ! 
while True:
    print('루프~ ')
    break
    
# -> 반복1번 실행후 루프 출력 ! 바로밑에 break를 만나 while문 종료
```

```python
number = 1
while True:
    print(number)
    number +=1
    if number > 3:
        break
```

```python
num = 0
while True:
    print(num)
    num +=1
    if num >= 13: 
        break
```

1. while 실습예제) 두개의 정수를 입력 받아서 더하는 코드를 작성하시오.
(단, 두개의 정수가 0이 들어올 때까지 반복한다.)


```python
# 1. 두개 정수 입력 먼저 받아보기

# 2.결과화면을 살펴보니 입력부분이 계속 반복이네요 ! 입력을 반복문 안에 넣어주기 ! 

# 3. 반복문안에서 입력이 0,0이 들어왔는지를 판별하는 조건식, 맞으면 멈추는 코드 작성

# 4. 0,0이 아니라면 두 정수의 합을 계산하는 연산코드 작성
```

```python
while True:
    num1 = int(input('정수입력1 >> '))
    num2 = int(input('정수입력2 >>'))
    
    if num1 ==0 and num2 ==0:
        break
    
    result = num1 + num2
    print('두 정수의 합:',result)
print('프로그램이 종료되었습니다.')
```

2. while실습예제) 다이어트 관리 프로그램 실습 
현재 몸무게와 목표몸무게를 입력 받고 주차 별 감량 몸무게를 입력 받으세요.  
목표몸무게를 달성하면 축하한다는 문구를 출력하고 입력을 멈추세요!


```python
cnt = 0    
now = int(input('현재 몸무게: '))
goal = int(input('목표 몸무게: '))
while True:
    cnt +=1
    lw = int(input(f'{cnt}주차 감량 몸무게: '))
    now -= lw
    if now <= goal:
        print(f'{now}kg 달성 ! 축하합니다 !! ')
        break
```

### for문

####  for - range 함수 사용 1
range(시작값,끝값,증가량)  
range(1,10,1) -> 1부터 9까지 1씩 증가 (시작값 포함, 끝값 포함 x)  
range(1,100,3) -> 1부터 99까지 3씩 증가  
range(10,1,-1) -> 10부터 2까지 1씩 감소(-1)

```python
range(1,10,1)
```

```python
for i in range(1,10,1):
    print(i)
```

```python
for i in range(1,10,1):
    print(i, end = ' ')
```

####  for - range 함수 사용 2
range(시작값,끝값,증가량)    
기본 설정 값 >>>>   range(0,끝값,1)  

```python
# range(시작, 끝) 2값이 들어가면 시작과 끝값을 의미 , 뒤의 증가값은 1로 생략 가능 
for i in range(3,10):
    print(i, end = ' ')
```

```python
# range(끝) 값이 1 들아가면 끝값 의미, 시작0, 증가량은 1로 생략 가능
for i in range(10):
    print(i, end = ' ')
```

#### for - range 함수 사용 3
range(len(list)) 사용  
특정 길이를 반복 횟수로 사용하는 방법  
리스트의 인덱스 사용 편의, 효율적 

```python
list1 =['햄버거','피자','핫도그']
for i in range(0,len(list1),1):
    print(i, list1[i])
```

```python
list1 =['햄버거','피자','핫도그']

# 시작값, 증가량 생략 가능 ! 왜냐 기본 값으로 설정할 거니까 굳이 안적어 줘도 됨
for i in range(len(list1)):
    print(i, list1[i])
```

2. for 문 실습 예제


```python
list2 = [4,5,2,1,99,15,2,7,27]
max_data = list2[0]

for i in list2 :
    if max_data<i :
        max_data = i
print(max_data)
```

5. for문 실습 예제


```python
num = int(input("정수를 입력해 주세요 !! "))
print(f"{num}의 약수 : ", end = " ")
for i in range(1,num+1):
    if num % i == 0:
        print(i, end = " ")
```

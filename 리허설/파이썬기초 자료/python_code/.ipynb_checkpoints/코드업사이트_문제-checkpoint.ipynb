```python
#  
#     
# a = b+c
import numpy as np
# q1.   2, (0,0) (1,0). (in)/(on)/(out) .
center_x,center_y = (0,0)
tree_x,tree_y = (3,0)
distance = np.sqrt((center_x-tree_x)**2+(center_y-tree_y)**2)
if distance == 2:
    print(distance,'(on).')
elif distance < 2:
    print(distance,'(in).')
else:
    print(distance,'(out).')
```

```python
# q2.   3, (1,1) (3,4). (in)/(on)/(out) .
center_x,center_y = (1,1)
tree_x,tree_y = (3,4)
distance = np.sqrt((center_x-tree_x)**2+(center_y-tree_y)**2)
if distance == 3:
    print('on')
elif distance < 3:
    print('in')
else:
    print('out')
```

```python
#1. 비밀번호 정보 설정, 수정
# - input() 받아 pw 변수에 저장
#2. 잠금 해제하기 비밀번호 입력
# - 비밀번호 입력 성공 > "잠금이 해제되었습니다. Unlock"(출력문 보여주기, 반복끝)
# - 비밀번호 입력 틀림 > 잘못된 비밀번호입니다. 다시 입력해주세요 >>> (다시반복)
```

```python
# 비밀번호 확인
def pw_check():
    print(f""" 
            {'='*30}
                  * Password Check *
            {'-'*30}
            l{' '*28}l
            l{' '*28}l
            l{'    비밀번호를 입력하세요!  l'}
            l    ---------------------   l
            l    l    _ _ _ _ _ _    l   l
            l    ---------------------   l 
            l{' '*28}l
            l{' '*28}l
            {'-'*30}
            {'='*30}
            """)
pw_check()
```

```python
# 잠금해제
def pw_success():
    print(f""" 
            {'='*30}
                    * Unlocked *
            {'-'*30}
            l{' '*28}l
            l{' '*28}l
            l{'          ᒄ₍⁽ˆ⁰ˆ⁾₎ᒃ♪♬        l'}
            l{' '*28}l
            l{'         welcome~~~♬       l'}
            l{' '*28}l
            {'-'*30}
            {'='*30}
            """)

def pw_success2():    
    print(f""" 
            {'='*30}
                    * Unlocked *
            {'-'*30}
            l{'           welcome~~~ ♥    l'}
            l{' '*28}l
            l{' '*28}l
            l{'         ก็็็็็็็็็็็็็  ʕ•͡ᴥ•ʔ  ก้้้้้้้้้้้        l'}
            l{'         welcome~~~♬       l'}
            l{' '*28}l
            {'-'*30}
            {'='*30}
            """)    
    
    
    
```

```python
# 잠금해제 실패
def pw_fail():
    print(f""" 
            {'='*30}
                    *  locked  *
            {'-'*30}
            l{' '*28}l
            l{' '*28}l
            l{'      잘못입력했습니다!     l'}
            l{' '*28}l
            l{' '*28}l
            {'-'*30}
            {'='*30}
            """)
```

```python
# 비밀번호 변수
pw = '12345*' # 여섯자리

pw_check()
pw_ipt = input('잠금상태 | 비밀번호를 눌러주세요>> ')
while True:
    if pw_ipt == pw:
        pw_success()
        print('잠금이 해제되었습니다!!')
        break
    else:
        pw_fail()
        pw_ipt = input('잘못된 비밀번호입니다. 다시 입력하여 주세요>> ')
        continue
```

```python
# 비밀번호 변수
pw = '12345*' # 여섯자리 
# 1. 1번을 눌렀을대 잠금 해제  
# 2. 2번을 눌렀을때 비밀 번호 설정
cnt = False
while True:
    print('--------메뉴를 선택하세요--------')
    choice = input('[1]잠금해제 [2]비밀번호 설정>> ')
    print()
    if choice =='1' and cnt:
        pw_check()
        pw_ipt = input('- 잠금상태 | 비밀번호를 눌러주세요>> ')
        while True:
            if pw_ipt == pw:
                pw_success2()
                print('- 잠금이 해제되었습니다!!')
                break
            else:
                pw_fail()
                pw_ipt = input('- 잘못된 비밀번호입니다. 다시 입력하여 주세요>> ')
                continue
        break
    elif choice == '2':
        cnt = True
        pw = input('- 새 비밀번호를 입력하세요>> ')
        print('- 새로운 비밀번호가 설정되었습니다!')
        print()
    else:
        print('- 잘못 선택하셨습니다!!')
        print()
```

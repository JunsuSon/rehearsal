딕셔너리(dictionary) for문 활용 예제

```python
score_dict = {'이름':['손지영','김은영','김운비','채수민','최성우'],
        'Python':[90,100,85,90,80],
        'Java':[85,80,100,95,85],
        'html/css':[75,70,90,80,90]}
score_dict
```

```python
#1. python 점수의 평균 구하기
py_avg = 0
py_avg = sum(score_dict['Python'])/5
print(py_avg)
```

```python
score_dict.keys()
```

```python
# 2. 각 과목의 평균 구하기
py_avg,j_avg,h_avg = 0,0,0
for i in score_dict.keys():
    if i =='Python':
        py_avg = sum(score_dict[i])/5
    elif i =='Java':
        j_avg = sum(score_dict[i])/5
    elif i =='html/css':
        h_avg = sum(score_dict[i])/5
print('Python 평균:',py_avg)
print('Java 평균:',j_avg)
print('html/css 평균:',h_avg)
```

```python
# 2. 각 과목의 평균 구하기
avg = 0
for i in list(score_dict.keys())[1:]:
    avg = sum(score_dict[i])/5
    print(f'{i} 평균:',avg)
```

```python
# 3. 손지영이 있으면 성적 전체 평균 구하기 
avg = 0
check = False
input_name = input('이름을 정확하게 기입하세요 >> ')
for i in range(len(score_dict['이름'])):
    if input_name == score_dict['이름'][i]:
        avg = (score_dict['Python'][i] + score_dict['Java'][i]+ score_dict['html/css'][i])/3
        print(f'{input_name}의 전 과목 성적 평균: {avg:.2f}')
        check = True

if not check : # 없었다 라는 의미
    print('검색 결과가 없습니다.')
```

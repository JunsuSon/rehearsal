```python
# 네임스페이스 (namespace, 이름 공간)
# 프로그래밍 언어에서 특정한 객체(object)를 이름(name)에 따라 구분할 수 있는 범위
# 전역 네임스페이스, 지역 네임스페이스 
x = 9
print('함수 밖:', x)
def temp():
    x = 10 # 함수 안에 있는 x변수
    print('함수 안:', x)
temp()
```

```python
# global 키워드 
x = 9
print('전역 변수:', x)
def temp():
    global x # x를 전역 변수로 만듬
    x = 10 # 함수 안에 있는 x변수
    print('전역 변수 개념으로 사용:', x)
temp()
print('전역 변수2:', x)
```

```python
# 함수 안에 내용 짜보기 : 어떤 수의 약수를 구하는 코드 
# 함수로 정의 
def divisor(num):
    for i in range(1,num+1):
        if num % i == 0:
            print(i, end = '\t')
divisor(10)
```

```python
def cal(num1, num2, op):
    # 함수안에서 사용할 매개변수명 num1, num2, op
    result = 0
    if op == "+":
        # + 연산 
        result =  num1 + num2
    elif op == '-':
        # 빼기 연산
        result =  num1 - num2
    return result
    
# 함수 호출(사용)
num1 = int(input('정수1>> '))
num2 = int(input('정수2>> '))
op = input('연산 입력(+,-)>> ')
result = cal(num1, num2, op)
print(result)
```

```python
# 함수 설명문
# 독스트링(document string: docstring)
# 기능 설명을 작성하는 부분
# 정의된 함수 내에서 shift + tab 누르면 설명 확인 가능
# 함수 내부 첫 줄에 개행을 포함한 긴 줄 문자열 기호(""" """)로 작성
def cal(num1, num2, op):
    """ 덧셈과 뺄셈을 계산하는 함수
    num1: int, num2: int, op: 연산자string """
    result = 0
    if op == "+":
        # + 연산 
        result =  num1 + num2
    elif op == '-':
        # 빼기 연산
        result =  num1 - num2
    return result
```

```python
cal() # shift + tab
```

```python
# 함수 return 키워드 정리 
# 1. return있는 함수의 반환값은 언제나 하나 
def add_sub(num1, num2):
    return num1+num2, num1-num2, num1-num2 # # packing : 포장 
```

```python
a,b,c = add_sub(10,7)
print(a)
print(b)
print(c)
# a,b,c 로 각각 따로 담을 수 있음 -> unpacking : 포장을 품
```

```python
# 지수 연산하는 함수 정의
def power_num(num , power = 2):
    return num ** power
```

```python
# 가변 매개변수(*)
def add(*num):
    total = 0
    for i in num:
        total += i # 1, 2, 3, 6, ...
    return total
    
add(1,2,3,6,7,8,5)
```

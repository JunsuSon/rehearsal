```python
# break: while문을 나가는 키워드
cnt = 1
while True:
    print(cnt, '파이썬이 쉬워요!')
    cnt += 1
    # if 문으로 조건 만들어보기 13번까지 출력 멈추기 
    if cnt == 14:
        break
```

```python
while True:
    num1 = int(input('첫 번째 정수 입력>> '))
    num2 = int(input('두 번째 정수 입력>> '))   
    # 조건
    if num1 == 0 and num2 == 0:
        print('프로그램이 종료되었습니다.')
        break
    print('두 정수의 합:', num1 + num2)
```

```python
# 랜덤한 정수형태 난수 생성하는 방법 
import random 
com_num = random.randint(1,50) # 끝값을 포함하는 함수 
com_num
num = 0  
cnt = 1
# 반복문 멈추는 키워드 : break 
while True:
    # 입력을 계속 
    num = int(input(f'{cnt} - 숫자를 입력하세요 >> '))
    cnt += 1 # cnt = cnt + 1
    if com_num > num:
        print(f'{num}보다 큰 수 입니다.')
    elif com_num < num:
        print(f'{num}보다 작은 수 입니다.')
    else:
        print('정답을 맞추셨습니다~~~!')
        break
```

```python
# for 문 돌리며 각 요소의 숫자를 확인하고
# 합격 불합격 프린트하기 
# * 몇 번째 학생인지 구하는 방법 1. 카운팅 2. 인덱스 번호 함수 
# 대상list.index(값)
score_list = [90,45,70,60,55]
cnt = 1
for score in score_list:
    # 합격여부 판별
    if score >=60:
        print(f'{cnt}번 학생은 합격입니다.')
    else:
        print(f'{cnt}번 학생은 불합격입니다.')
    cnt += 1
```

```python
for score in score_list:
    # 합격여부 판별
    if score >=60:
        print(f'{score_list.index(score)+1}번 학생은 합격입니다.')
    else:
        print(f'{score_list.index(score)+1}번 학생은 불합격입니다.')
```

#### 이중 리스트 반복문 활용하기

```python
frd_list = [['강련',28],['건욱',29],['선웅',26],
           ['주섭',27],['인우',26],['나행',25]]
print('리스트의 길이는?', len(frd_list))
```

```python
# 건욱의 나이를 출력 (접근)
# 인덱싱 
frd_list[1][1]

# 인우씨의 나이 접근
frd_list[4][1]

# 나행의 나이 접근
frd_list[-1][1]
```

```python
# 모든 사람의 나이 데이터만 출력하시오 
# 반복문 for 

# 주섭 데이터가 있다면 주섭 나이 출력하기 
for data in frd_list:
    # 주섭의 이름이 있는지
    if data[0] == '주섭':
        print(data[1])
```

```python
#빈 리스트 2개 
name_list = [] # 이름만 담아주기
age_list = []  # 나이 데이터만 담아주기 

# 리스트 맨 뒤로 추가하는 함수 : 대상리스트.append(값) 
for name, age in frd_list:
    name_list.append(name)
    age_list.append(age)

# name_list, age_list 출력해보기 
print(name_list)
print(age_list)
```

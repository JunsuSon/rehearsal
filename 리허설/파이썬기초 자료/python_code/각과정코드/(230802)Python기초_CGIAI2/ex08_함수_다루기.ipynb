```python
#     :      
#   
def divisor(num):
    for i in range(1,num+1):
        if num % i == 0:
            print(i, end = '\t')
divisor(10)
```

```python
def cal(num1, num2, op):
    #    num1, num2, op
    result = 0
    if op == "+":
        # +  
        result =  num1 + num2
    elif op == '-':
        #  
        result =  num1 - num2
    return result
    
#  ()
num1 = int(input('1>> '))
num2 = int(input('2>> '))
op = input(' (+,-)>> ')
result = cal(num1, num2, op)
print(result)
```

```python
#  
# (document string: docstring)
#    
#    shift + tab    
#          (""" """) 
def cal(num1, num2, op):
    """    
    num1: int, num2: int, op: string """
    result = 0
    if op == "+":
        # +  
        result =  num1 + num2
    elif op == '-':
        #  
        result =  num1 - num2
    return result
```

```python
cal() # shift + tab
```

```python
#  (*)
def add(*num):
    total = 0
    for i in num:
        total += i # 1, 2, 3, 6, ...
    return total
    
add(1,2,3,6,7,8,5)
```

```python
# list.index() ->    
# 
# 
#  ~ (   index()  )
#  ~ 
lst = ['','','','','']
search = input('  >> ')
if search in lst:
    #  ~ 
    print(f'{search}  {lst.index(search)+1} .')
else:
    print(f'{search}   .')
```

```python
# .count()
#   
s = 'Hi, My name is Mihee'
s = s.lower()
#    
search = input('  : ')
cnt = 0
# 
if search in s:
    cnt = s.count(search)
    print(f"'{search}' {cnt}  .")
else:
    print(f"'{search}'  .")
    
```

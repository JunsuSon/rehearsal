### 산술연산자

```python
# 나누기, 나머지, 나누기(몫) 구하기
# num1 변수 10 대입
# num2 변수 7 대입
num1 = 10
num2 = 7

print('나누기 결과:',num1/num2)
print('나누기 몫 결과:',num1//num2)
print('나머지 결과:',num1%num2)
```

```python
# 문자열 더하기 
# str1  '안녕' 대입하기
# str2 '하세요' 대입하기
# 두개의 변수 더하기 출력
str1 = '안녕'
str2 = '하세요'
print(str1 + str2)
```

```python
str3 = '10'
str4 = '7'
# 두변수를 더하기
print(str3 + str4)
```

```python
# num1 과 str4 는 더해질까?
# 숫자 + 문자열 더해지지 않음
# num1 = 10, str4 = '7'
str(num1) + str4
```

### 복합대입 연산자

```python
# = 대입 연산자
# += 더하고 대입
# -= 빼고 대입
# *= 곱하고 대입
# /= 나누고 대입
# %= 나머지값 대입 
# +=  기준으로 설명
a = 3
b = 2
a += b # a = a + b
print(a)
```

```python
a,b = b,a
print(a,b) 
# 파이썬 치환 : 새로운 임의의 변수 생성 없이 교환 가능  
```

### 비교 연산자

```python
# a > b : a가 b보다 크다
# a >= b : a가 b보다 크거나 같다
# a < b : b가 a보다 크다
# a <= b : b가 a보다 크거나 같다
# a == b: a와 b가 같다
# a != b : a와 b는 다르다 (같지 않다)
```

```python
# random 모듈(라이브러리)
import random
a = random.randint(1,10)
b = random.randint(1,10)

print(a < b)
print(a > b)
print(a == b)
print(a != b)
print(a, b)
```

### 논리 연산자

```python
# and : 논리1 and 논리2
# a> b and a != b =>True , 1
# a< b and a != b => False, 0

# or : 논리1 or 논리2
# True or False => True, 1
# True or True => True,1

# not : not 논리 
# 논리를 뒤집는 것 
```

```python
print(True and False)
print(True or False)
print(not False) # if 조건식 부분에서 많이 활용 
```

```python
print(2>1 and 2==1) 
print(2==2 or 2<1)
print(not 2<1)
```

### 멤버연산자

```python
# 조건문에 자주 활용이 됨
# in : 찾고자 하는 값(x)이 포함되어 있으면 True
# not in : 찾고자 하는 값(x)이 포함되지 않으면 True
# x in 문자열
'a' in 'aaaabbbbccc' # 뒷 문장안에 a 가 있어? True
# x not in 문자열
'd' not in 'aaaabbbbccc' # 뒷 문장안에 d가 없지? True
```

```python
'g' in ['a','b','c','d']
```

### 삼항 연산자 

```python
# True결과 if 조건식 else False결과
score = 80 # 60점 이상 
'합격' if score >=60 else '불합격'
```

```python
num = int(input('정수 입력>> '))
result = '짝수' if num%2==0 else '홀수'
print(f'{num}은(는) {result}입니다.')
# 삼항 연산자도 연산자! 결과를 변수에 대입해서 활용 가능 
```

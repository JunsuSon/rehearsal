```python
# pickle -> pkl
# [], () 자료구조를 python 형식 그대로 저장하게 하는 확장자 pkl
import pickle
# file = open('경로 및 파일명', '읽기 모드/쓰기모드')
file = open('./data/movie_review.pkl', 'rb')
data = pickle.load(file)
data # [리뷰문장내용, 긍정1/부정0]
file.close() # open 함수 file 닫기 
```

```python
# 1. 전체 리뷰 개수 확인하기 
len(data)
```

```python
# 2. 리뷰 길이 평균 값
# 리뷰 접근 
sum_len = 0
for rv in data:
    sum_len += len(rv[0])  # 각 문장의 길이  sum_len = sum_len + len(rv[0])
print(sum_len/len(data))
```

```python
# 1 긍정
# 0 부정 
# 0,1 숫자의미가 없는 데이터(크고작은 의미가 없는 것 )
neg,pos = 0,0
for cont, sent in data:
    if sent == 0:
        neg +=1 # 카운팅
    else: 
        pos +=1 # 긍정 카운팅 
print('긍정:', pos)
print('부정:', neg)
```

```python
max_len = len(data[0][0])
max_cont = ''
for cont, sent in data:
    if len(cont) > max_len:
        max_len = len(cont)
        max_cont = cont
print('리뷰내용 최대길이:', max_len)
print(max_cont)
```

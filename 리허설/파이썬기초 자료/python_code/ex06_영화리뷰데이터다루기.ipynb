```python
# 파일 불러오기
import pickle 
with open('../data/movie_review.pkl', 'rb') as f:
    print(type(f))
    data = pickle.load(f)
type(data[0][0])
```

```python
# 1. 전체 리뷰 개수 확인 
len(data)
```

```python
# 2. 리뷰 길이 평균
tt_len = 0
for i in data:
    tt_len += len(i[0])
print('리뷰 길이 평균:',tt_len/len(data))
```

```python
#3. 긍정리뷰 1 / 부정리뷰 0 개수
p = 0 #positive
n = 0 #negative
for i in data:
    if i[1] == 1: #긍정 count
        p += 1
    else: #부정 count
        n +=1
print('긍정: ', p)
print('부정: ', n)
```

```python
#4. 리뷰 길이 최대값 / 리뷰 내용 출력
max_num = len(data[0][0]) # 첫번째 문장 길이가 최대가 임시로 설정
max_content = ''
for i in data:
    if max_num < len(i[0]):
        max_num = len(i[0])
        max_content = i[0]
        
print(f'리뷰 최대 길이는 \'{max_num}\'입니다.')
print(max_content)
```

```python
# 5. 'ㅋ'가 포함된 리뷰 개수 
cnt_k = 0
for i in data:
    if 'ㅋ' in i[0]:
         cnt_k += 1
print(cnt_k)
```

```python
# 6. '검색데이터'가 포함된 리뷰 개수
input_word = input('검색 데이터 입력  >> ')
cnt_w = 0
for i in data:
    if input_word in i[0]:
        cnt_w +=1
print(cnt_w)
```
